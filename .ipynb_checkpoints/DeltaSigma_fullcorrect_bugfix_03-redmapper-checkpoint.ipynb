```python
fig = plt.figure()
ax = fig.add_subplot(111)


ax.errorbar(ACP.rr, ACP.rr*ACP.dst, ACP.rr*ACP.dst_err, capsize=4, label="with Sel. resp.")
ax.errorbar(ACP_nometa.rr, ACP.rr*ACP_nometa.dst, ACP.rr*ACP_nometa.dst_err, capsize=4, label="without Sel. resp")

ax.set_xscale("log")
ax.set_yscale("log")

ax.set_xlabel("R [Mpc]")
   
ax.set_title("redMaPPer clusters $45 < \lambda < 50$ and $0.2 < z < 0.35$")    
    
ax.legend()
ax.set_ylim(5, 100)
ax.set_ylabel("$R\cdot\Delta\Sigma$ [$M_\odot  / pc$]")
```
